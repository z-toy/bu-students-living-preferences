Is East or West Best?


For years, this ongoing debate among Boston University students has left us wondering which location is the best to live at on campus. For the purpose of this project, three major questions and hypotheses were raised to help answer this question.

1. Is there a correlation between a student’s major and how important it is for them to live near their classes? And how does this relate to where they currently live?
*Hypothesis: STEM majors tend to prefer to live near their classes compared to non-STEM majors.*

2. Is there a correlation between Greek life membership and how important it is to live near Allston?
*Hypothesis: Greek life members tend to prefer to live near Allston.*

3. Do underclassmen prefer to be closer to the city? In other words, is there a correlation between graduation year and how important it is for them to live near the city?
*Hypothesis: Underclassmen tend to prefer to live closer to the city compared to upperclassmen.*

Data Collection 

To gather our data, acquaintances completed a survey. Respondents were with asked background questions on their gender, graduating year, current college of enrollment, where they currently live, if they are a member Greek life, and if they are a varsity or club athlete. Respondents were then asked to answer the following questions:

On a scale of 1-10 (1 being the least, 10 being the most), how important is being close to:
1. FitRec
2. Allston (Greek Life)
3. Classes
4. The City (Newbury, Prudential, etc.)
The results of the survey were then compiled into a .csv file.


Data Visualization

```python
import pandas as pd
from google.colab import files
import io

uploaded = files.upload() 
```

```python
df = pd.read_csv(io.BytesIO(uploaded['survey.csv']))
df
```

1. Is there a correlation between a student’s major and how important it is for them to live near their classes? And how does this relate to where they currently live?

```python
import seaborn as sns
barplot = sns.barplot(x='classes', y='major', data=df)
barplot.set(xlabel='importance of class distance',ylabel='major')
```

```python
violin = sns.catplot(x='classes', y='major', kind="violin", inner=None, data=df)
sns.swarmplot(x='classes', y='major', color="k", size=4, data=df, ax=violin.ax)
violin.set(xlabel='importance of class distance',ylabel='major')
```

```python
boxplot = sns.boxplot(y=df['major'], x=df['classes'])
boxplot.set(xlabel='importance of class distance',ylabel='major')
```

```python
swarm = sns.swarmplot(x='classes', y='major', data=df, size=10)
swarm.set(xlabel='importance of class distance',ylabel='major')
```

```python
cat = sns.catplot(x='classes', y='major',
                  col='location', data=df, kind="swarm",
                  height=5, aspect=.3);
cat.set(xlabel='importance of dist.',ylabel='major')
```

2. Is there a correlation between Greek life membership and how important it is to live near Allston?

```python
barplot = sns.barplot(x ='allston', y='greek_life', data = df)
barplot.set(xlabel='importance of living near allston',ylabel='greek life membership')
```

```python
boxplot = sns.boxplot(y=df['greek_life'], x=df['allston'])
boxplot.set(xlabel='importance of living near allston',ylabel='greek life membership')
```

```python
catplot = sns.catplot(x='greek_life', y='allston',
            kind="violin", data=df, cut=0)
catplot.set(ylabel='importance of living near allston',xlabel='greek life membership')
```

3. Do underclassmen prefer to be closer to the city? In other words, is there a correlation between graduation year and how important it is for them to live near the city?

```python
barplot = sns.barplot(x ='grad_year', y='city', data = df)
barplot.set(ylabel='importance of living near city',xlabel='graduation year')
```

```python
boxplot = sns.boxplot(y=df['city'], x=df['grad_year'])
boxplot.set(xlabel='grad year',ylabel='importance of living near city')
```

```python
catplot = sns.catplot(data=df, x='grad_year', y='city', kind='violin', cut=0)
catplot.set(ylabel='importance of living near city',xlabel='graduation year')
```

```python
sns.set_theme(style="whitegrid") 
cmap = sns.cubehelix_palette(rot=-.2, as_cmap=True)
g = sns.relplot(
    data=df,
    x="grad_year", y="city",
    hue="city", size="location",
    palette=cmap, sizes=(10, 100)
)

g.ax.xaxis.grid(False)
g.ax.yaxis.grid(False)
```

```python
sns.relplot(x="grad_year", y="city", kind="line", data=df);
```

```python
x = df["grad_year"]
y = df["city"]

import matplotlib.pyplot as plt

plt.plot(x,y,"o")
```

```python
sns.jointplot(data=df, x="grad_year", y="city", hue="location")
```
